### Bitwise Operations
- XOR is exclusive or operator



- `5 ^ 5 = 0` and `0 ^ 3 = 0`
- So, exclusive or the expected numbers and the array will yield the missing number in the array

```python
def missing_number(nums):
    res = len(nums)

    for i in range(len(nums)):
        res ^= i ^ nums[i]
    
    return res
```

### Gauss' Formula
- Expected sum of $\frac{n}{2} (n+1)$
- `missing = expected_sum - actual_sum`

```python
def missing_number(nums):
    expected = len(nums) * (len(nums) + 1) // 2
    actual = sum(nums)
    return expected - actual
```
